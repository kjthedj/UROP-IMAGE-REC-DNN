## Fraunhofer diffraction

# 1D Fraunhofer diffraction
The Fraunhofer diffraction pattern for a single slit can be represented as a one-dimensional matrix or graph. The intensity distribution I(Î¸) in the diffraction pattern can be calculated using the following formula for a single slit:

Screenshot 2023-10-28 162248.png

To create a matrix representing this pattern, you can calculate the intensity at different angles Î¸ and plot them.

This code will generate a plot of the Fraunhofer diffraction pattern for a single slit. One can adjust the parameters a and Î» to see how they affect the pattern. Note that this is a one-dimensional representation, and for a more detailed two-dimensional representation, you'd need to consider a rectangular aperture and perform a 2D Fourier transform.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
a = 0.1  # Width of the slit (in arbitrary units)
Î» = 0.5  # Wavelength of light (in the same units)
I0 = 1.0  # Maximum intensity (in arbitrary units)

# Define the range of angles
theta = np.linspace(-np.pi/2, np.pi/2, 1000)

# Calculate the intensity distribution
I = I0 * (np.sin(np.pi * a * np.sin(theta) / Î») / (np.pi * a * np.sin(theta) / Î»))**2

# Create the plot
plt.plot(theta, I)
plt.xlabel('Angle (radians)')
plt.ylabel('Intensity')
plt.title('Fraunhofer Diffraction Pattern for a Single Slit')
plt.show()
```

## 2D Fraunhofer Diffraction: Heat Map

Screenshot 2023-10-28 162051.png
This modified code calculates the intensity distribution in 2D by using a 2D grid of angles, Theta_x and Theta_y, and then visualizes the 2D diffraction pattern with a 2D heatmap using imshow. The resulting plot will show the diffraction pattern in both the horizontal and vertical directions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
a_x = 0.3  # Width of the slit in the horizontal direction (in arbitrary units)
a_y = 0.3  # Width of the slit in the vertical direction (in arbitrary units)
Î» = 0.3    # Wavelength of light (in the same units)
I0 = 1.0   # Maximum intensity (in arbitrary units)

# Define the range of angles
theta_x = np.linspace(-np.pi/2, np.pi/2, 1000)
theta_y = np.linspace(-np.pi/2, np.pi/2, 1000)

# Create a grid of theta_x and theta_y values
Theta_x, Theta_y = np.meshgrid(theta_x, theta_y)

# Calculate the intensity distribution in 2D
intensity = I0 * (np.sin(np.pi * a_x * np.sin(Theta_x) / Î») / (np.pi * a_x * np.sin(Theta_x) / Î»))**2 * (np.sin(np.pi * a_y * np.sin(Theta_y) / Î») / (np.pi * a_y * np.sin(Theta_y) / Î»))**2

# Create the 2D plot
plt.figure(figsize=(8, 6))
plt.imshow(intensity, extent=[theta_x.min(), theta_x.max(), theta_y.min(), theta_y.max()], cmap='hot', aspect='auto')
plt.colorbar(label='Intensity')
plt.xlabel('Angle (radians) - Horizontal')
plt.ylabel('Angle (radians) - Vertical')
plt.title('2D Fraunhofer Diffraction Pattern for a 2D Slit')
plt.show()
```

## 2D Fraunhofer Diffraction: Matrix Derivations

```python
import numpy as np

# Define parameters
a_x = 0.1  # Width of the slit in the horizontal direction (in arbitrary units)
a_y = 0.1  # Width of the slit in the vertical direction (in arbitrary units)
Î» = 0.5    # Wavelength of light (in the same units)
I0 = 0.9   # Maximum intensity (in arbitrary units)

# Define the range of angles
theta_x = np.linspace(-np.pi/2, np.pi/2, 10)
theta_y = np.linspace(-np.pi/2, np.pi/2, 10)#value of 10 is currently arbitrarily plugged in

# Create a grid of theta_x and theta_y values
Theta_x, Theta_y = np.meshgrid(theta_x, theta_y)

# Calculate the intensity distribution in 2D
intensity = I0 * (np.sin(np.pi * a_x * np.sin(Theta_x) / Î») / (np.pi * a_x * np.sin(Theta_x) / Î»))**2 * (np.sin(np.pi * a_y * np.sin(Theta_y) / Î») / (np.pi * a_y * np.sin(Theta_y) / Î»))**2

# Print the entire intensity matrix
print(intensity)
```

# Fresnal Diffraction

Fresnel diffraction refers to the diffraction of light waves when they pass through small apertures or encounter obstacles. It's often described using Fresnel diffraction integrals, which can be calculated numerically using various programming languages. Below is a Python code example for simulating Fresnel diffraction patterns using the Fresnel-Kirchhoff diffraction integral:
Screenshot 2023-10-28 161915.png



The codes below are for different sorts of aperature such as gaussian, circular and rectangular. In each case the matrix for intensity is printed right after each heat maps

Fresnal diffraction with gaussian aperature

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

# Constants
wavelength = 0.0005  # Wavelength of light (in meters)
k = 2 * np.pi / wavelength  # Wave number

# Define the aperture dimensions and create a grid
aperture_size = 0.01  # Aperture size (in meters)
num_points = 40  # Number of grid points
x = np.linspace(-aperture_size / 2, aperture_size / 2, num_points)
y = np.linspace(-aperture_size / 2, aperture_size / 2, num_points)
x, y = np.meshgrid(x, y)

# Define a more complex aperture function (example: Gaussian)
sigma = 0.002  # Standard deviation for Gaussian
aperture = np.exp(-x**2 / (2 * sigma**2) - y**2 / (2 * sigma**2))

# Define the observation plane coordinates and create a grid
observation_plane_size = 0.01  # Observation plane size (in meters)
x_observation = np.linspace(-observation_plane_size / 2, observation_plane_size / 2, num_points)
y_observation = np.linspace(-observation_plane_size / 2, observation_plane_size / 2, num_points)
x_observation, y_observation = np.meshgrid(x_observation, y_observation)

# Initialize the intensity pattern
intensity_pattern = np.zeros_like(x_observation, dtype=float)

# Calculate the 2D Fresnel diffraction integral (optimized)
for i in range(num_points):
    for j in range(num_points):
        dx = x_observation[i, j] - x
        dy = y_observation[i, j] - y
        r = np.sqrt(dx**2 + dy**2)

        #avoid division by 0 and handle the invalid values
        epsilon = 1e-10
        r = np.where(r<epsilon,epsilon,r)


        integral = simps(simps(aperture * np.exp(1j * k * r) / r, x[0], x[-1]), y[:, 0], y[:, -1])

        #handle invalid integrals
        if np.isnan(integral):
          integral = 0.0

        intensity_pattern[i, j] = np.abs(integral)**2

# Rescale the intensity pattern for visualization
intensity_pattern = intensity_pattern / np.max(intensity_pattern)

# Plot the intensity pattern with the 'hot' colormap
plt.imshow(intensity_pattern, cmap='hot', extent=(-observation_plane_size / 2, observation_plane_size / 2, -observation_plane_size / 2, observation_plane_size / 2))
plt.title('2D Fresnel Diffraction Pattern')
plt.colorbar()
plt.show()

#print out the matrix in each case
print(intensity_pattern)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

# Constants
wavelength = 0.0005  # Wavelength of light (in meters)
k = 2 * np.pi / wavelength  # Wave number

# Define the aperture dimensions and create a grid
aperture_size = 0.01  # Aperture size (in meters)
num_points = 5  # Number of grid points
x = np.linspace(-aperture_size / 2, aperture_size / 2, num_points)
y = np.linspace(-aperture_size / 2, aperture_size / 2, num_points)
x, y = np.meshgrid(x, y)

# Define a rectangular aperture
rect_width = 0.004  # Width of the rectangular aperture
rect_height = 0.002  # Height of the rectangular aperture
aperture = np.where((np.abs(x) <= rect_width / 2) & (np.abs(y) <= rect_height / 2), 1.0, 0.0)

# Define the observation plane coordinates and create a grid
observation_plane_size = 0.01  # Observation plane size (in meters)
x_observation = np.linspace(-observation_plane_size / 2, observation_plane_size / 2, num_points)
y_observation = np.linspace(-observation_plane_size / 2, observation_plane_size / 2, num_points)
x_observation, y_observation = np.meshgrid(x_observation, y_observation)

# Initialize the intensity pattern
intensity_pattern = np.zeros_like(x_observation, dtype=float)

# Calculate the 2D Fresnel diffraction integral (optimized)
for i in range(num_points):
    for j in range(num_points):
        dx = x_observation[i, j] - x
        dy = y_observation[i, j] - y
        r = np.sqrt(dx**2 + dy**2)

        # Avoid division by zero and handle invalid values
        epsilon = 1e-10
        r = np.where(r < epsilon, epsilon, r)

        integral = simps(simps(aperture * np.exp(1j * k * r) / r, x[0], x[-1]), y[:, 0], y[:, -1])

        # Handle invalid values (nan)
        if np.isnan(integral):
            integral = 0.0

        intensity_pattern[i, j] = np.abs(integral)**2

# Rescale the intensity pattern for visualization
intensity_pattern = intensity_pattern / np.max(intensity_pattern)

# Plot the intensity pattern with the 'hot' colormap
plt.imshow(intensity_pattern, cmap='hot', extent=(-observation_plane_size / 2, observation_plane_size / 2, -observation_plane_size / 2, observation_plane_size / 2))
plt.title('2D Fresnel Diffraction Pattern')
plt.colorbar()
plt.show()

#print the matrix derivation
print(intensity_pattern)
```
